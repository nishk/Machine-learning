```python
#Source - https://www.kaggle.com/datasets/karthickveerakumar/salary-data-simple-linear-regression

import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd

input_data = '/Users/anishkirloskar/Downloads/Machine-learning/2022Practice/SalaryLinearReg/Salary_Data.csv'
df = pd.read_csv(input_data)
print(df.head())

x = df.iloc[1:,0]
y = df.iloc[1:,1]

#print(x.shape)
#print(y.shape)
```

```python
def Forward(x):
    return w*x

def Loss(x,y):
    y_pred = Forward(x)
    return (y_pred - y)*(y_pred - y)

w_list = []
loss_list = []
w = 0.0

for w in np.arange(13000,13500,10):
    print("w=",round(w,2))
    l_sum=0
    for x_val,y_val in zip(x,y):
        y_pred = Forward(x_val)
        l = Loss(x_val, y_val)
        l_sum += l
        print("\t", round(x,2), round(y,2), round(y_pred,2), round(l_sum/3,2))
    print("MSE=",round(l_sum/3,2))
    w_list.append(w)
    loss_list.append(l_sum/3)
    
print("Minimum loss is:",min(loss_list))
print("Corresponding w is:", w_list[loss_list.index(min(loss_list))])
w_min = w_list[loss_list.index(min(loss_list))]
y_pred = w_min*np.asarray(x)


f1 = plt.figure(1)
plt.plot(w_list,loss_list)
plt.ylabel('Loss')
plt.xlabel('w')

f2 = plt.figure(2)
plt.plot(x, y, label='Actual Values', color='Red')
plt.plot(x, y_pred, label='Predicted Values', color='Green')
plt.xlabel('Hours')
plt.ylabel('Salary')
plt.legend()
#print(loss_list)
```
