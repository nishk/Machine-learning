```python
#headers
import numpy as np
import matplotlib.pyplot as plt
import math

%precision %.2f

x_data = [1.0,2.0,3.0]
y_data = [3.0,5.0,7.0]

#Start with random prediction
w=1.0

#Model Function for forward pass y=w.x
def forward(x):
    return w*x

#Define Loss function loss = (y-y_pred)^2
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y)*(y_pred - y)

w_list=[]
mse_list=[]
for w in np.arange(0.0,4.1,0.1):
    print("w=",round(w,2))
    l_sum=0
    for x_val,y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val,y_val)
        l_sum += l
        print("\t",x_val,y_val,y_pred_val,l)
    print("MSE=", round(l_sum/3,2))
    w_list.append(w)
    mse_list.append(l_sum/3)
    
print("min MSE=",min(mse_list))
print("Corresponding w=", w_list[mse_list.index(min(mse_list))])

w_min = w_list[mse_list.index(min(mse_list))]
print("w_min=", w_min)
y_pred_val = w_min*np.asarray(x_data)
print("Predicted y values=",y_pred_val)

f1 = plt.figure(1)
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')

f2 = plt.figure(2)
plt.plot(x_data, y_data, label ='Actual Values', color='red')
plt.plot(x_data,y_pred_val, label ='Predicted Values', color='green')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.legend()
plt.show()

```
