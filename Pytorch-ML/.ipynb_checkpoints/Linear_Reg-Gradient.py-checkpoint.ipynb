```python
##This is an implementation of simple linear regression with gradient descent
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x_data = [2.0,3.0,4.0,5.0]
y_data = [7.0,8.0,9.0,10.0]
w = 1.0
w_list = []
mse_list = []

def forward(x,w):
    return w*x

def loss(x,y,w):
    y_pred = forward(x,w)
    return (y_pred -y)**2

def gradient(x,y,w):
    return 2*x*(x*w-y)

print("Before training", forward(6,w))

#Training
for epoch in range(100):
    for x_val,y_val in zip(x_data,y_data):
        grad =gradient(x_val,y_val,w)
        w = w-0.01*grad
        print("\tgrad: ",x_val,y_val,grad)
        l = loss(x_val,y_val,w)
    print("progress:", epoch,"w=",w,"loss=",l)
    
#After training
print("predict after training", forward(6,w))
```
