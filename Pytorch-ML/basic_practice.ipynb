```python
#Simple Univariate Linear regression

import torch
import numpy as np
import matplotlib.pyplot as plt


x_data = [1.0,2.0,3.0]
y_data = [10.0,12.0,14.0]
w_list=[]
mse_list=[]
w=0.0

def forward(x):
    return w*x

def loss(x_val,y_val):
    y_pred = forward(x_val)
    return ((y_pred - y_val)**2)

for w in np.arange(0.0,10.1,0.1):
    print("w=",w)
    for x_val,y_val in zip(x_data,y_data):
        l_sum =0
        y_pred_val = forward(x_val)
        l = loss(x_val,y_val)
        l_sum+=l
        print("\t",x_val,y_val,y_pred_val,l)
    print("MSE=",l_sum/3)
    w_list.append(w)
    mse_list.append(l_sum/3)

print("min Error is:", min(mse_list))
print("Equivalent w is:",w_list[mse_list.index(min(mse_list))])
plt.plot(w_list,mse_list)
plt.ylabel('MSE')
plt.xlabel('w')
plt.show()
```
