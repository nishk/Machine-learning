```python
#Simple Univariate Linear regression

import torch
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0,2.0,3.0,4.0]
y_data = [3.0,5.0,8.0,11.0]
m = len(y_data)
w = 1
w_list = []
mse_list = []

#Forward function for our model -> y = w*x 
def forward(x):
    return w*x

#Loss Function
def loss(x,y):
    y_pred = forward(x)
    return (y_pred -y)**2

for w in np.arange(0.0, 5.1, 0.2):
    print("w=",w)
    l_sum=0
    for x_val, y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val,y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum/m)
    w_list.append(w)
    mse_list.append(l_sum/m)
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show();
#min_w = min(w_list)
#print (min_w)
index_min_mse = mse_list.index(min(mse_list)) 
print ("Min error is:",mse_list[index_min_mse])
print ("The corresponding w is:",w_list[index_min_mse])

y_model_pred = w_list[index_min_mse]*np.asarray(x_data)
print(y_model_pred, y_data)

plt.plot(x_data,y_data, label='Actual_model',color='red')
plt.plot(x_data,y_model_pred,color='green', label='Predicted model')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show();
```
