```python
import torch
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0,2.0,3.0,4.0]
y_data = [6.0,7.0,8.0,9.0]
m = len(y_data)
w = 1
w_list = []
mse_list = []

def forward(x):
    return w*x

def loss(x,y):
    y_pred = forward(x)
    return (y_pred -y)**2

for w in np.arange(0.0, 8.1, 0.2):
    print("w=",w)
    l_sum=0
    for x_val, y_val in zip(x_data,y_data):
        y_pred_val = forward(x_val)
        l = loss(x_val,y_val)
        l_sum += l
        print("\t", x_val, y_val, y_pred_val, l)
    print("MSE=", l_sum/m)
    w_list.append(w)
    mse_list.append(l_sum/m)
plt.plot(w_list,mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show();
```
