```python
import torch
import numpy as np
import matplotlib.pyplot as plt


x = torch.tensor(10.0,requires_grad=True)
b = torch.tensor(8.0,requires_grad=True)

y = x**2 + b*x + 10

# Calculating gradient/derivative of y w.r.t x
y.backward()

#grad is property/attribute of x, hence grad is correct & not grad()
print(x.grad)
print(b.grad)

#https://stackoverflow.com/questions/48001598/why-is-zero-grad-needed-for-optimization
x.grad.zero_()

print(x.grad)
```

```python
a = torch.linspace(-10,10,10,requires_grad=True)
B = a**2
b = torch.sum(B)

b.backward()

plt.plot(a.detach().numpy(),B.detach().numpy(),label='Function')
plt.plot(a.detach().numpy(),a.grad.detach().numpy(),label='Derivative')
plt.legend()
```
